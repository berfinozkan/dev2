```python
##palindrom konrolü
def palindromKontrol(string):
    string=string.lower()
    string=string.replace(" ", "")
    tersi=string[::-1]
    return string== tersi
    
print(palindromKontrol("A man a plan a canal Panama"))
print(palindromKontrol("hello"))
```

```python
##İki Sayının En Büyük Ortak Böleni (EBOB)
def ebob(a, b):
    while b:
        a, b = b, a % b
    return a

print(ebob(50, 48))
```

```python
## Anagram Kontrolü
def anagramKontrol(str1, str2):
    str1 = str1.lower().replace(" ", "")
    str2 = str2.lower().replace(" ", "")

    if len(str1) != len(str2):
        return False
    for harf in str1:
        if str1.count(harf) != str2.count(harf):
            return False
    else:
        return True

print(anagramKontrol("listen", "silent"))
```

```python
##eksik sayi bulma
def eksikSayi(liste):
    eksikler=[]
    N = len(liste) + 1  
    for i in range(1, N + 1):
        if i not in liste:
            eksikler.append(i)
    return eksikler
print(eksikSayi([1, 2, 5]))  
```

```python
##Listeyi Belirli Bir Değere Göre Döndürme 
def saga_dondur(liste, k):
    while k > len(liste):
        k -= len(liste)
    return liste[-k:] + liste[:-k]

print(saga_dondur([1, 2, 3, 4, 5], 3))
```

```python
class Ogrenci:
    def __init__(self,ad,numara,notlar=None):
        if notlar is None:
            notlar = [] 
        self.ad=ad
        self.numara=numara
        self.notlar=notlar

    def notEkle(self,note):
        if 0<= note <= 100:
            self.notlar.append(note)
        else:
            print("Not 0 ile 100 arasında olmalıdır!")

    def ortalamaHesapla(self):
        if len(self.notlar) == 0:
            return 0
        return sum(self.notlar) / len(self.notlar)


    def bilgileriGoster(self):
        ort = self.ortalamaHesapla()
        print(" Öğrenci: " + self.ad +" Numara: "+str(self.numara)+ " Notlar: " +str(self.notlar) + " Ortalama: "+ str(round(ort,2)))
```

```python
o1 = Ogrenci("Ali", 123, [90, 80, 70])
o1.bilgileriGoster()
o1.notEkle(95)
o1.bilgileriGoster()
```

```python
class Kitap:
    def __init__(self, baslik, yazar, sayfa_sayisi):
        self.baslik = baslik
        self.yazar = yazar

        try:
            if sayfa_sayisi < 0:
                print("Hata: Sayfa sayısı negatif olamaz! 0 olarak ayarlandı.")
                self.sayfa_sayisi = 0
            else:
                self.sayfa_sayisi = sayfa_sayisi
        except:
            print("Hata: Sayfa sayısı bir sayı olmalıdır!")
            self.sayfa_sayisi = 0

        self.okunan_sayfa = 0


    def oku(self, sayfa):
        """Belirtilen kadar sayfa oku."""
        try:
            if sayfa < 0:
                print("Negatif sayfa okunamaz!")
                return
            
            if self.okunan_sayfa + sayfa > self.sayfa_sayisi:
                print("Bu kadar sayfa yok! Son sayfaya kadar okundu.")
                self.okunan_sayfa = self.sayfa_sayisi
            else:
                self.okunan_sayfa += sayfa
        except:
            print("Hata: Okunacak sayfa sayısı sayı olmalıdır!")

    def sayfa_atla(self, sayfa):
        """Belirtilen kadar sayfa ileri atla."""
        try:
            if sayfa < 0:
                print("Negatif sayfa atlanamaz!")
                return

            if self.okunan_sayfa + sayfa > self.sayfa_sayisi:
                print("Bu kadar ileri atlanamaz, kitap sonuna atlandı.")
                self.okunan_sayfa = self.sayfa_sayisi
            else:
                self.okunan_sayfa += sayfa
        except:
            print("Hata: Sayfa atlama değeri sayı olmalıdır!")

    def bilgileri_goster(self):
        print(f"Kitap: {self.baslik} - {self.yazar}")
        print(f"Toplam Sayfa: {self.sayfa_sayisi}")
        print(f"Okunan Sayfa: {self.okunan_sayfa}")
```

```python
k = Kitap("Madonna", "Ali", 300)
k.oku(50)
k.sayfa_atla(20)
k.oku(20)
k.bilgileri_goster()
```
